```python
import mysql.connector
mydb = mysql.connector.connect(host= 'localhost', passwd='Haris12@sql',user='root',database='teacher')
cursor = mydb.cursor()

def findData():
    if cursor.rowcount>0:
        print ("Data Updated Successfully...")
    else:
        print ("No Data Found.")
while True:
    Action = int(input('''   1. For Insert Data
    2. For Display all record
    3. For Display one Record
    4. For update data
    5. For Delete data'''))
    if Action == 1 :
        Emplo_ID = int(input('Enter your ID'))
        Name = input('Enter your name')
        Gender = input('Enter your gender')
        City = input('Enter your city')
        Age = int(input('Enter your Age'))

        query = "insert into task values({},'{}','{}','{}',{})".format(Emplo_ID,Name,Gender,City,Age)
        cursor.execute(query)
        mydb.commit()
        print('Data inserted')

    elif  Action == 2 :
        cursor.execute("select * from task")
        for i in cursor:
            print(i)
            
    elif  Action == 3 :
        Empl_id = int(input('Enter Id'))
        query = "select * from task where Emplo_ID = {}".format(Empl_id)
        cursor.execute(query)
        for x in cursor:
            print(f'\n{x}')
            
    elif  Action == 4 :
        Empl_id = int(input('Enter Id'))
        
        Action = int(input('''1. For update Name/Gender/City
    2. For update age'''))
        if Action == 1:
            wish = input("Name/Gender/City")
            inpu = input(f'Enter {wish}')
            query = "update task set {}='{}' where Emplo_ID={}".format(wish,inpu,Empl_id)
            cursor.execute(query)
            mydb.commit()
            findData()
        elif Action == 2:
            age= input('Enter Age')
            query = "update task set Age={} where Emplo_ID={}".format(age,Empl_id)
            cursor.execute(query)
            mydb.commit()
            findData()
                
    elif  Action == 5 :
        Empl_id = int(input('Enter Id'))
        query = "DELETE from task where Emplo_ID = {}".format(Empl_id)
        cursor.execute(query)
        mydb.commit()
        findData()
```
